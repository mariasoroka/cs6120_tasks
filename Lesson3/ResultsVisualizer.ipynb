```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import re
```

```python
benchmarks_names = []
old_results = []
new_results = []

with open('/Users/ms3663/Courses/CS6120/Assignments/Lesson3/results_float.csv', mode ='r') as file:
    csvFile = csv.reader(file)
    for i, lines in enumerate(csvFile):
        if i == 0:
            continue
        if i%2 == 0:
            name = lines[0]
            new_name = re.sub('[^a-zA-Z]+', '', name)
            benchmarks_names.append(new_name)
            old_results.append(int(lines[2]))
        else:
            new_results.append(int(lines[2]))
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_yscale('log')
ax.bar(np.arange(len(benchmarks_names)), old_results, color = 'teal', width = 0.35, label = 'Baseline')
ax.bar(np.arange(len(benchmarks_names)) + 0.4, new_results, color = 'olive', width = 0.35, label = 'Optimized')
ax.set_xticks(np.arange(len(benchmarks_names)))
ax.set_xticklabels(benchmarks_names, rotation=90)
ax.legend()
fig.show()

new_results = np.array(new_results)
old_results = np.array(old_results)

speedup = old_results/new_results
print(np.mean(speedup))
```
